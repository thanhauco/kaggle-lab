# Forecasting with Facebook Prophet\n

```python
from prophet import Prophet\nimport pandas as pd\n
```

```python
df = pd.read_csv('../data/train.csv')\n# Prepare for Prophet\nprophet_df = df.groupby('date')['sales'].sum().reset_index()\nprophet_df.columns = ['ds', 'y']\n
```

## Training\n

```python
m = Prophet(daily_seasonality=True, yearly_seasonality=True)\nm.fit(prophet_df)\n
```

## Prediction\n

```python
future = m.make_future_dataframe(periods=30)\nforecast = m.predict(future)\nm.plot(forecast)\nplt.show()\n
```

```python
m.plot_components(forecast)\nplt.show()\n
```
